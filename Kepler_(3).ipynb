```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Dias = np.array([88.0 , 224.7 , 365.3 , 687.0, 4331.8 , 10760.0, 30684.0, 60188.3 ])
Distancia = np.array([36 , 67.25, 93, 141.75, 483.8, 887.97, 1764.5, 2791.05])


plt.scatter(Dias,Distancia)
plt.show()
```

```python
from scipy import interpolate
```

```python
spline = interpolate.InterpolatedUnivariateSpline(Dias, Distancia)
x1 = np.linspace(min(Dias),max(Dias)) 


y1 = spline(x1) 

plt.plot(Dias, Distancia, 'ro', x1, y1)

plt.axis([min(x1)-1, max(x1)+1, min(y1)-1, max(y1)+1])
```
